```python
import os
import cv2
import pandas as pd
import numpy as np
import boto3
import json
from credentials.amazon_credentials import amazon_key
import re
import matplotlib.pyplot as plt
import os

def detect_faces(photo, region='us-west-2'):
    # Create a Rekognition client
    session = boto3.Session(region_name=region, aws_access_key_id=amazon_key['AWS_ACCESS_KEY'], aws_secret_access_key=amazon_key['AWS_SECRET_KEY'])
    client = session.client('rekognition', region_name=region)

    # Load image from local file system
    with open(photo, 'rb') as image_file:
        image_bytes = image_file.read()

    # Call Rekognition to detect faces
    response = client.detect_faces(
        Image={'Bytes': image_bytes},
        Attributes=['ALL']
    )

    print('Detected faces for ' + photo)
    for faceDetail in response['FaceDetails']: # works on single face images only
        return faceDetail    

def flatten_dict(d, parent_key='', sep='_'):
    items = {}
    if d:
        for key, value in d.items():
            new_key = f"{parent_key}{sep}{key}" if parent_key else key
            if isinstance(value, dict):
                items.update(flatten_dict(value, new_key, sep=sep))
            elif isinstance(value, list):
                for i, sub_dict in enumerate(value):
                    sub_dict_items = flatten_dict(sub_dict, new_key, sep=sep)
                    for sub_key, sub_value in sub_dict_items.items():
                        if sub_key.split('_')[-1] == 'Type':
                            s = sub_value
                        elif sub_key == 'Confidence' or 'X' or 'Y':
                            items[f"{new_key}_{s}_{sub_key.split('_')[-1]}"] = sub_value
            else:
                items[new_key] = value
    else:
        items = {key: None for key in ['BoundingBox_Width', 'BoundingBox_Height', 'BoundingBox_Left', 
                                       'BoundingBox_Top', 'AgeRange_Low', 'AgeRange_High', 'Smile_Value', 
                                       'Smile_Confidence', 'Eyeglasses_Value', 'Eyeglasses_Confidence', 
                                       'Sunglasses_Value', 'Sunglasses_Confidence', 'Gender_Value', 
                                       'Gender_Confidence', 'Beard_Value', 'Beard_Confidence', 'Mustache_Value', 
                                       'Mustache_Confidence', 'EyesOpen_Value', 'EyesOpen_Confidence', 
                                       'MouthOpen_Value', 'MouthOpen_Confidence', 'Emotions_SURPRISED_Confidence', 
                                       'Emotions_FEAR_Confidence', 'Emotions_ANGRY_Confidence', 
                                       'Emotions_CONFUSED_Confidence', 'Emotions_DISGUSTED_Confidence', 
                                       'Emotions_HAPPY_Confidence', 'Emotions_CALM_Confidence', 
                                       'Emotions_SAD_Confidence', 'Landmarks_eyeLeft_X', 'Landmarks_eyeLeft_Y', 
                                       'Landmarks_eyeRight_X', 'Landmarks_eyeRight_Y', 'Landmarks_mouthLeft_X', 
                                       'Landmarks_mouthLeft_Y', 'Landmarks_mouthRight_X', 'Landmarks_mouthRight_Y',
                                       'Landmarks_nose_X', 'Landmarks_nose_Y', 'Landmarks_leftEyeBrowLeft_X', 
                                       'Landmarks_leftEyeBrowLeft_Y', 'Landmarks_leftEyeBrowRight_X', 
                                       'Landmarks_leftEyeBrowRight_Y', 'Landmarks_leftEyeBrowUp_X', 
                                       'Landmarks_leftEyeBrowUp_Y', 'Landmarks_rightEyeBrowLeft_X', 
                                       'Landmarks_rightEyeBrowLeft_Y', 'Landmarks_rightEyeBrowRight_X', 
                                       'Landmarks_rightEyeBrowRight_Y', 'Landmarks_rightEyeBrowUp_X', 
                                       'Landmarks_rightEyeBrowUp_Y', 'Landmarks_leftEyeLeft_X', 
                                       'Landmarks_leftEyeLeft_Y', 'Landmarks_leftEyeRight_X', 
                                       'Landmarks_leftEyeRight_Y', 'Landmarks_leftEyeUp_X', 'Landmarks_leftEyeUp_Y',
                                       'Landmarks_leftEyeDown_X', 'Landmarks_leftEyeDown_Y', 
                                       'Landmarks_rightEyeLeft_X', 'Landmarks_rightEyeLeft_Y', 
                                       'Landmarks_rightEyeRight_X', 'Landmarks_rightEyeRight_Y', 
                                       'Landmarks_rightEyeUp_X', 'Landmarks_rightEyeUp_Y', 
                                       'Landmarks_rightEyeDown_X', 'Landmarks_rightEyeDown_Y', 
                                       'Landmarks_noseLeft_X', 'Landmarks_noseLeft_Y', 
                                       'Landmarks_noseRight_X', 'Landmarks_noseRight_Y', 'Landmarks_mouthUp_X', 
                                       'Landmarks_mouthUp_Y', 'Landmarks_mouthDown_X', 'Landmarks_mouthDown_Y', 
                                       'Landmarks_leftPupil_X', 'Landmarks_leftPupil_Y', 'Landmarks_rightPupil_X', 
                                       'Landmarks_rightPupil_Y', 'Landmarks_upperJawlineLeft_X', 
                                       'Landmarks_upperJawlineLeft_Y', 'Landmarks_midJawlineLeft_X', 
                                       'Landmarks_midJawlineLeft_Y', 'Landmarks_chinBottom_X', 'Landmarks_chinBottom_Y', 
                                       'Landmarks_midJawlineRight_X', 'Landmarks_midJawlineRight_Y', 
                                       'Landmarks_upperJawlineRight_X', 
                                       'Landmarks_upperJawlineRight_Y', 'Pose_Roll', 'Pose_Yaw', 'Pose_Pitch', 'Quality_Brightness', 
                                       'Quality_Sharpness', 'Confidence', 'FaceOccluded_Value', 'FaceOccluded_Confidence', 'EyeDirection_Yaw', 
                                       'EyeDirection_Pitch', 'EyeDirection_Confidence']}
    return items
```

Function local_directory: Outputs a combined dataframe for all images in a directory

```python
def local_directory(dir_name):
    df = pd.DataFrame()
    json_file = {}
    for file in os.listdir(dir_name):
        photo = os.path.join(dir_name, file)
        data = detect_faces(photo)
        flat_data = flatten_dict(data)
        df = pd.concat([df, pd.DataFrame([flat_data], index=[file])])
    return df

#TODO: Add Function for Online Image Repository
```

Function for for formatting and saving the amazon csv file (for 1 emotion)

```python
def value_and_conf(x):
    # Define a regex pattern to match column names
    pattern = r'(.*)_Value'  # This pattern matches any column name ending with '_Value'

    # Iterate over columns in the DataFrame
    for column in x.index:
        match = re.match(pattern, column)
        if match:
            name = match.group(1)  # Extract the name from the column name
            confidence_column = f"{name}_Confidence"  # Construct the corresponding confidence column name
            
            # Perform the calculation if both value and confidence columns exist
            if x[name + '_Value']:
                x[name] = 100 + x[confidence_column]
            else:
                x[name] = 100 - x[confidence_column]
    return x

def drop_unnecessary(df):
    visited_col = set()
    pattern = r'(.*)_Value'

    # Iterate over columns in the DataFrame
    for column in df.columns.tolist():
        match = re.match(pattern, column)
        if match:
            name = match.group(1)
            visited_col.add(name + '_Value')
            visited_col.add(name + '_Confidence')  # Corrected from undefined 'confidence_column'
    
    # Drop the identified columns
    df.drop(columns=visited_col, inplace=True, errors='ignore')
    return df
```

```python
directory = 'rafdb_15k/DATASET/train'

# Get a list of all CSV files in the directory
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]

# Create an empty list to store individual DataFrames
dfs = []

# Iterate over each CSV file
for file in csv_files:
    # Read the CSV file into a DataFrame and append it to the list
    df = pd.read_csv(os.path.join(directory, file))
    df['label'] = file.split('_')[0]
    dfs.append(df)

# Concatenate all DataFrames in the list along the rows (axis=0)
concatenated_df = pd.concat(dfs, axis=0)
concatenated_df = concatenated_df.apply(value_and_conf, axis=1)
concatenated_df = drop_unnecessary(concatenated_df)
concatenated_df.to_csv(directory+'_concat.csv')
```

```python
def show_NaN_images(df):
    # Filter the DataFrame to retain only rows with NaN values
    df_nan = df[df.isna().any(axis=1)]
    
    # Calculate the number of rows needed based on the number of NaN images
    num_images = len(df_nan)
    num_rows = (num_images + 9) // 10
    
    fig, axes = plt.subplots(num_rows, 10, figsize=(20, num_rows * 2))
    
    # Iterate over each row in the filtered DataFrame
    for idx, row in df_nan.reset_index().iterrows():
        # Calculate the position in the grid
        row_index = idx // 10
        col_index = idx % 10
        
        # Construct the file path for the image
        file_path = os.path.join('rafdb_15k/DATASET', f"{row['Image_Name'].split('_')[0]}/{row['label']}/", row['Image_Name'])
        
        # Load and display the image
        image = plt.imread(file_path)
        axes[row_index, col_index].imshow(image)
        axes[row_index, col_index].axis('off')  # Turn off axis
        axes[row_index, col_index].set_title(f"{row['Image_Name']}")  # Set title with index
        if idx == (num_rows * 10) - 1:  # Break loop if we have displayed 10 images per row
            break
    
    # Hide empty subplots
    for i in range(num_images, num_rows * 10):
        axes.flatten()[i].axis('off')
    
    plt.tight_layout()
    plt.show()

#show_NaN_images(train)
```

```python
import os
import matplotlib.pyplot as plt
import altair as alt
import pandas as pd

def show_random_amazon(df, index=0):
    row = df.iloc[0]
    file_path = os.path.join('rafdb_15k/DATASET', f"{row['Image_Name'].split('_')[0]}/{row['label']}/", row['Image_Name'])
    image = plt.imread(file_path)
    plt.imshow(image)
    plt.axis('off')
    
    img_h, img_w, img_ch = image.shape
    
    #Draw a rectangle with row['BoundingBox_Width','BoundingBox_Height','BoundingBox_Left','BoundingBox_Top']
    width = row['BoundingBox_Width'] * img_w
    height = row['BoundingBox_Height'] * img_h
    left = row['BoundingBox_Left'] * img_w
    top = row['BoundingBox_Top'] * img_h
    plt.gca().add_patch(plt.Rectangle((left, top), width, height, linewidth=1, edgecolor='r', facecolor='none'))
    
    pattern = r'(.*)_X'  # This pattern matches any column name ending with '_Value'

    # Prepare data for scatter plot
    scatter_data = []
    for column in row.index:
        match = re.match(pattern, column)
        if match:
            name = match.group(1)
            scatter_data.append({'x': row[name+'_X']*img_w, 'y': row[name+'_Y']*img_h, 'label': name})
    
    scatter_df = pd.DataFrame(scatter_data)
    
    # Create Altair scatter plot
    scatter_chart = alt.Chart(scatter_df).mark_circle(size=60).encode(
        x='x',
        y='y',
        color='label:N',
        tooltip=['label']
    ).interactive()
    
    plt.show()
    return scatter_chart

# Assuming 'test' is your DataFrame
show_random_amazon(test)
```

```python
def openface_to_dataframe(dir_name='/Users/jiheeyou/subjective-output/openface-results/train_set'):
    df = pd.DataFrame()
    for file in os.listdir(dir_name):
        if file.endswith('.csv'):
            file_path = os.path.join(dir_name, file)
            new_df = pd.read_csv(file_path)
            new_df = new_df.sort_values(by=[' confidence'], ascending=False).head(1)
            new_df.index = [file]
            df = pd.concat([df, new_df], axis=0)
        else:
            continue
    return df
openface_df = openface_to_dataframe()
```
