# Extracting Amazon Face Attributes

Import statements (NOTE: you need your amazon key inside credentials/amazon_credentials) \
Function detect_faces: Connects to Amazon \
Function flatten_dict: Flattens retrieved dictionary

```python
import os
import cv2
import pandas as pd
import numpy as np
import boto3
import json
from credentials.amazon_credentials import amazon_key

def detect_faces(photo, region='us-west-2'):
    # Create a Rekognition client
    session = boto3.Session(region_name=region, aws_access_key_id=amazon_key['AWS_ACCESS_KEY'], aws_secret_access_key=amazon_key['AWS_SECRET_KEY'])
    client = session.client('rekognition', region_name=region)

    # Load image from local file system
    with open(photo, 'rb') as image_file:
        image_bytes = image_file.read()

    # Call Rekognition to detect faces
    response = client.detect_faces(
        Image={'Bytes': image_bytes},
        Attributes=['ALL']
    )

    print('Detected faces for ' + photo)
    for faceDetail in response['FaceDetails']: # works on single face images only
        return faceDetail    

def flatten_dict(d, parent_key='', sep='_'):
    items = {}
    for key, value in d.items():
        new_key = f"{parent_key}{sep}{key}" if parent_key else key
        if isinstance(value, dict):
            items.update(flatten_dict(value, new_key, sep=sep))
        elif isinstance(value, list):
            for i, sub_dict in enumerate(value):
                sub_dict_items = flatten_dict(sub_dict, new_key, sep=sep)
                for sub_key, sub_value in sub_dict_items.items():
                    if sub_key.split('_')[-1] == 'Type':
                        s = sub_value
                    elif sub_key == 'Confidence' or 'X' or 'Y':
                        items[f"{new_key}_{s}_{sub_key.split('_')[-1]}"] = sub_value
        else:
            items[new_key] = value
    return items
```

Function local_directory: Outputs a combined dataframe for all images in a directory

```python
def local_directory(dir_name):
    df = pd.DataFrame()
    json_file = {}
    for file in os.listdir(dir_name):
        photo = os.path.join(dir_name, file)
        data = detect_faces(photo)
        flat_data = flatten_dict(data)
        df = pd.concat([df, pd.DataFrame([flat_data], index=[file])])
    return df

#TODO: Add Function for Online Image Repository
```

```python
def openface_to_dataframe(dir_name='/Users/jiheeyou/subjective-output/openface-results/train_set'):
    df = pd.DataFrame()
    for file in os.listdir(dir_name):
        if file.endswith('.csv'):
            file_path = os.path.join(dir_name, file)
            new_df = pd.read_csv(file_path)
            new_df = new_df.sort_values(by=[' confidence'], ascending=False).head(1)
            new_df.index = [file]
            df = pd.concat([df, new_df], axis=0)
        else:
            continue
    return df
openface_df = openface_to_dataframe()
```
